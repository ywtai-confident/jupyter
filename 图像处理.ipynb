```python
# 图像锐化：增强图像清晰度
def image_sharp_sobel(image_arr):
    src = cv2.GaussianBlur(image_arr, (3, 3), 0)
    grad_x = cv2.Sobel(src, -1, 1, 0, ksize=3)
    abs_grad_x = cv2.convertScaleAbs(grad_x)
    grad_y = cv2.Sobel(src, -1, 0, 1, ksize=3)
    abs_grad_y = cv2.convertScaleAbs(grad_y)
    grad  = cv2.addWeighted(abs_grad_x, 0.5, abs_grad_y, 0.5, 0)
    return grad

def image_sharp_laplacian(image_arr):
    src = cv2.GaussianBlur(image_arr, (3, 3), 0)
    dst = cv2.Laplacian(src, -1, ksize = 3)
    grad = cv2.convertScaleAbs(dst)
    return grad
```

```python
image_abnormal_sharpen = image_sharp_laplacian(image_abnormal)
abnormal_res = image_abnormal - image_abnormal_sharpen
# 像素值大于150的点，采用周围的点进行插值
noise_idx = np.argwhere(abnormal_res > 150)
noise_xy = {(x[1], x[0]) for x in noise_idx[:, :2]}
empty_xy = []
for ni in noise_xy:
    # 获取周围8领域的像素坐标（除开异常点）
    x, y = ni[0], ni[1]
    idx8_list = [(x-1, y), (x-1, y-1), (x-1, y+1), (x, y-1), (x, y+1), (x+1, y), (x+1, y+1), (x+1, y-1)]
    # 去掉异常点
    idx8_list = [xi for xi in idx8_list if xi not in noise_xy and 0 <= xi[0] < abnormal_res.shape[1]
                 and 0 <= xi[1] < abnormal_res.shape[0]]
    if not idx8_list:
        empty_xy.append(ni)
        continue
    # 采用周围像素均值的中位数对应的下标像素进行填充
    pix8_avg = [abnormal_res[xi[1], xi[0]].mean() for xi in idx8_list]
    idx_choose = idx8_list[np.argmin(np.abs(pix8_avg - np.median(pix8_avg)))]
    for i in range(abnormal_res.shape[2]):
        abnormal_res[y, x, i] = abnormal_res[idx_choose[1], idx_choose[0], i]
cv2.imwrite('D:\\demo\\GasMeterNumber\\images\\test.jpg', abnormal_res)
cv2.imshow('ori', image_abnormal)
cv2.imshow('laplacian', image_abnormal_sharpen)
cv2.imshow('sharpened', abnormal_res)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
